```python
import folium

```

```python
m = folium.Map(location=[15.3583,75.1194])
```

```python
m
```

```python
tooltip = "Click me!"
```

```python
folium.Marker(
    [ 15.3583,75.1194 ], popup="hola", tooltip=tooltip
).add_to(m)
folium.Marker(
    [ 14.3,75.12 ], popup="hello", tooltip=tooltip
).add_to(m)
```

```python
m
```

```python
m.save("index.html")
```
